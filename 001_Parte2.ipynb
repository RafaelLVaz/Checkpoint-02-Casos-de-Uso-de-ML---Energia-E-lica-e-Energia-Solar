## Carregamento de dados

```python
import pandas as pd

df = pd.read_csv('/content/smart_grid_stability_augmented.csv')

display(df.head())
```

## Preparação dos dados


```python
X = df.drop(['stabf', 'stab'], axis=1)
y = df['stabf']

display(X.head())
display(y.head())
display(X.info())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

display(X_scaled[:5])
```

## Divisão dos dados


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## Treinamento dos modelos

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

dt_model = DecisionTreeClassifier()
knn_model = KNeighborsClassifier()
lr_model = LogisticRegression()

dt_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)
```

## Avaliação dos modelos

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

dt_pred = dt_model.predict(X_test)
knn_pred = knn_model.predict(X_test)
lr_pred = lr_model.predict(X_test)

dt_accuracy = accuracy_score(y_test, dt_pred)
dt_confusion = confusion_matrix(y_test, dt_pred)
dt_f1 = f1_score(y_test, dt_pred, pos_label='unstable')

knn_accuracy = accuracy_score(y_test, knn_pred)
knn_confusion = confusion_matrix(y_test, knn_pred)
knn_f1 = f1_score(y_test, knn_pred, pos_label='unstable')

lr_accuracy = accuracy_score(y_test, lr_pred)
lr_confusion = confusion_matrix(y_test, lr_pred)
lr_f1 = f1_score(y_test, lr_pred, pos_label='unstable')

print("Desempenho do modelo de árvore de decisão:")
print(f"Acurácia: {dt_accuracy:.4f}")
print("Matriz de confusão:")
display(dt_confusion)
print(f"F1-score: {dt_f1:.4f}")
print("-" * 30)

print("Desempenho do modelo KNN:")
print(f"Acurácia: {knn_accuracy:.4f}")
print("Matriz de confusão:")
display(knn_confusion)
print(f"F1-score: {knn_f1:.4f}")
print("-" * 30)

print("Desempenho do modelo de regressão logística:")
print(f"Acurácia: {lr_accuracy:.4f}")
print("Matriz de confusão:")
display(lr_confusion)
print(f"F1-score: {lr_f1:.4f}")
```

### **Discussão**

Com base nas novas métricas de avaliação, o modelo de Árvore de Decisão continua sendo o mais confiável para a tarefa de detectar instabilidade na rede elétrica. Ele alcançou a maior acurácia (0.8975) e o **maior F1-score (0.9196) entre os três modelos testados.

O F1-score é uma métrica crucial neste contexto, pois oferece um balanço entre precisão e sensibilidade (recall), sendo fundamental para avaliar a capacidade do modelo em identificar corretamente os casos de instabilidade, minimizando erros como falsos positivos e falsos negativos.

O modelo KNN apresentou um desempenho competitivo e muito próximo, com um F1-score de 0.9143, mas a ligeira superioridade da Árvore de Decisão a torna a escolha preferencial. A Regressão Logística, por sua vez, demonstrou um desempenho consideravelmente inferior, com um F1-score de 0.8550, mostrando-se menos adequada para este problema de classificação.
